```python
import numpy as np

A = np.random.randint(1, 10, (3, 5))
B = np.random.randint(1, 10, (5, 2))
print("1) A * B (3x5 * 5x2):\n", np.dot(A, B))

M3 = np.random.randint(1, 10, (5, 3))
V = np.array([1, 2, 3])
print("\n2) Matrix (5x3) * Vector (3):\n", np.dot(M3, V))

coeffs = np.array([[3, 2], [1, 4]])
vals = np.array([12, 14])
print("\n3) Solution to system [3x+2y=12, x+4y=14]:\n", np.linalg.solve(coeffs, vals))

M5 = np.random.randint(1, 5, (5, 5))
det_val = np.linalg.det(M5)
print(f"\n4) Determinant of 5x5 matrix: {det_val:.4f}")

print("\n5) Inverse matrix:\n", np.linalg.inv(M5))
print("\nTranspose matrix:\n", M5.T)

eig_vals = np.linalg.eigvals(M5)
eig_prod = np.prod(eig_vals)
print(f"\n6) Product of eigenvalues: {eig_prod}")
print(f"Comparison: Det = {det_val:.4f}, Product = {eig_prod:.4f}")
print(f"Difference: {abs(det_val - eig_prod)}")
```
