```python
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt

import sympy as sp

x = sp.Symbol('x')
y = sp.Symbol('y')

f = x**3 - 3*x

df = sp.diff(f, x)
itg = sp.integrate(f, x)

print("--- Аналитический вид ---")
print(f"Исходная функция f(x) = {f}")
print(f"Производная f'(x) = {df}")
print(f"Интеграл F(x) = {itg}")

print("\n--- Решение нелинейного уравнения ---")
equation = sp.Eq(x**2 - 4, 0)
roots = sp.solve(equation, x)
print(f"Корни x^2 - 4 = 0: {roots}")

print("\n--- Решение системы уравнений ---")
system = [
    sp.Eq(x**2 + y**2, 5),
    sp.Eq(y, x + 1)
]
sys_roots = sp.solve(system, (x, y))
print(f"Решения системы: {sys_roots}")

print("\n--- Построение графиков ---")
from sympy.plotting import plot
p = plot(f, df, (x, -3, 3),
         title="f(x) - синий, f'(x) - красный",
         legend=True, show=False)

p[0].line_color = 'blue'
p[1].line_color = 'red'

p.show()
```
