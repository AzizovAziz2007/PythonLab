```python
atm_storage = {
    1000: 10,  
    500: 20,   
    100: 50,  
    50: 100,   
    10: 200  
}

try:
    amount = int(input("Введите сумму для выдачи: "))
    original_amount = amount

    result_list = []

    nominals = sorted(atm_storage.keys(), reverse=True)

    for nominal in nominals:
        count_needed = amount // nominal
        
        available = atm_storage[nominal]

        count_to_give = min(count_needed, available)

        if count_to_give > 0:
            result_list.append(f"{count_to_give} * {nominal}")

            amount -= count_to_give * nominal

            atm_storage[nominal] -= count_to_give
            
    if amount == 0:
        print("Выдача:", " + ".join(result_list))
    else:
        print("Операция не может быть выполнена!")
        
except ValueError:
    print("Пожалуйста, введите корректное число.")
```
