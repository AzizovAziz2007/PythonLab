```python
text = input("Введите текст: ")

words = text.split()

long_words = []
mid_words = []
short_words = []

for word in words:
    clean_word = word.strip(".,!?;:")
    lenght = len(clean_word)

    if(lenght > 7):
        long_words.append(clean_word)
    elif 4 <= lenght <= 7:
        mid_words.append(clean_word)
    else:
        short_words.append(clean_word)

print("Слова длиннее 7 символов:", long_words)
print("Слова от 4 до 7 символов:", mid_words)
print("Остальные слова:", short_words)
```
