```python
try:
    value = float(input("Введите вещественное число: "))

    if value < 0:
        raise ValueError("Некорректный формат!")

    rubles = int(value)

    kopecks = round((value-rubles)*100)

    print(f"{rubles} руб. {kopecks} коп.")
except ValueError:
    print("Некорректный формат!")
```
