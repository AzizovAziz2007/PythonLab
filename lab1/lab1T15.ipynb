```python
def pre_process(a=0.97):
    def decorator(func):
        def wrapper(s):
            s_new = s.copy()
            
            for i in range(1, len(s_new)):
                s_new[i] = s_new[i] - a * s_new[i-1]
            
            return func(s_new)
        return wrapper
    return decorator

@pre_process(a=0.93)
def plot_signal(s):
    for sample in s:
        print(round(sample, 2)) 

test_signal = [10, 20, 30, 40, 50]
plot_signal(test_signal)
```
