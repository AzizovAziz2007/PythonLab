```python
import random

n = random.randint(1, 10000)
print(f"Сгенерировано число n = {n}")

my_list = [random.randint(0, 100) for _ in range(n)]

target_size = 1
while target_size < n:
    target_size *= 2

print(f"Ближайшая степень двойки: {target_size}")

zeros_to_add = target_size - n
print(f"Нужно добавить нулей: {zeros_to_add}")

my_list += [0] * zeros_to_add

print(f"Итоговый размер массива: {my_list}")
```
