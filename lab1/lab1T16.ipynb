```python
import random
import itertools
from datetime import datetime, timedelta

teams = [
    "Реал Мадрид", "Барселона", "Бавария", "Ювентус",
    "Манчестер Сити", "Ливерпуль", "ПСЖ", "Челси",
    "Арсенал", "Милан", "Аякс", "Боруссия",
    "Порту", "Бенфика", "Шахтер", "Зенит"
]

random.shuffle(teams)

groups = [teams[i:i+4] for i in range(0, 16, 4)]

print("--- СОСТАВ ГРУПП ---")
for i, group in enumerate(groups, 1):
    print(f"Группа {i}: {', '.join(group)}")
print("\n--- РАСПИСАНИЕ ИГР ---")

game_date = datetime(2026, 9, 14, 22, 45)

all_group_matches = []
for group in groups:
    all_group_matches.append(list(itertools.combinations(group, 2)))

for tour in range(6):
    date_str = game_date.strftime("%d/%m/%Y, %H:%M")
    print(f"\nИгровой день: {date_str}")
    
    for g_idx in range(4):
        match = all_group_matches[g_idx][tour]
        print(f"  Группа {g_idx+1}: {match[0]} - {match[1]}")
    
    game_date += timedelta(weeks=2)
```
