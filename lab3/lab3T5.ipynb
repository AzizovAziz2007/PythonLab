```python
import sys
import re
from PyQt6.QtWidgets import (QApplication, QMainWindow, QWidget, QVBoxLayout,
                             QHBoxLayout, QLabel, QLineEdit, QCheckBox,
                             QSpinBox, QRadioButton, QPushButton, QButtonGroup)

class StringFormatter:
    def __init__(self, separators=None):
        self.separators = separators if separators else " "

    def _split_to_words(self, s):
        pattern = "[" + re.escape(self.separators) + "]+"
        return [w for w in re.split(pattern, s) if w]

    def remove_short_words(self, s, n):
        words = self._split_to_words(s)
        return " ".join([w for w in words if len(w) >= n])

    def replace_digits(self, s):
        return re.sub(r'\d', '*', s)

    def insert_spaces(self, s):
        return " ".join(list(s.replace(" ", "")))

    def sort_by_size(self, s):
        words = self._split_to_words(s)
        return " ".join(sorted(words, key=len))

    def sort_lexicographical(self, s):
        words = self._split_to_words(s)
        return " ".join(sorted(words))

class FormatterGUI(QMainWindow):
    def __init__(self):
        super().__init__()
        self.formatter = StringFormatter()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('StringFormatter Demo')
        self.setFixedSize(500, 400)

        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        main_layout = QVBoxLayout(central_widget)

        input_layout = QHBoxLayout()
        input_layout.addWidget(QLabel("Строка:"))
        self.input_field = QLineEdit()
        input_layout.addWidget(self.input_field)
        main_layout.addLayout(input_layout)

        short_layout = QHBoxLayout()
        self.cb_remove_short = QCheckBox("Удалить слова размером меньше")
        self.spin_n = QSpinBox()
        self.spin_n.setValue(5)
        short_layout.addWidget(self.cb_remove_short)
        short_layout.addWidget(self.spin_n)
        short_layout.addWidget(QLabel("букв"))
        short_layout.addStretch()
        main_layout.addLayout(short_layout)

        self.cb_replace_digits = QCheckBox("Заменить все цифры на *")
        main_layout.addWidget(self.cb_replace_digits)

        self.cb_insert_spaces = QCheckBox("Вставить по пробелу между символами")
        main_layout.addWidget(self.cb_insert_spaces)

        self.cb_sort = QCheckBox("Сортировать слова в строке")
        main_layout.addWidget(self.cb_sort)

        sort_group_layout = QVBoxLayout()
        sort_group_layout.setContentsMargins(30, 0, 0, 0)  # Отступ вправо
        self.rb_size = QRadioButton("По размеру")
        self.rb_lex = QRadioButton("Лексикографически")
        self.rb_size.setChecked(True)

        self.sort_group = QButtonGroup()
        self.sort_group.addButton(self.rb_size)
        self.sort_group.addButton(self.rb_lex)

        sort_group_layout.addWidget(self.rb_size)
        sort_group_layout.addWidget(self.rb_lex)
        main_layout.addLayout(sort_group_layout)

        self.btn_format = QPushButton("Форматировать")
        self.btn_format.clicked.connect(self.apply_formatting)
        main_layout.addWidget(self.btn_format)

        result_layout = QHBoxLayout()
        result_layout.addWidget(QLabel("Результат:"))
        self.result_field = QLineEdit()
        self.result_field.setReadOnly(True)  # Только для чтения
        result_layout.addWidget(self.result_field)
        main_layout.addLayout(result_layout)

    def apply_formatting(self):
        text = self.input_field.text()

        if self.cb_remove_short.isChecked():
            text = self.formatter.remove_short_words(text, self.spin_n.value())

        if self.cb_replace_digits.isChecked():
            text = self.formatter.replace_digits(text)

        if self.cb_insert_spaces.isChecked():
            text = self.formatter.insert_spaces(text)

        if self.cb_sort.isChecked():
            if self.rb_size.isChecked():
                text = self.formatter.sort_by_size(text)
            else:
                text = self.formatter.sort_lexicographical(text)

        self.result_field.setText(text)


if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = FormatterGUI()
    window.show()
    sys.exit(app.exec())
```
