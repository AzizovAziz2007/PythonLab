```python
class Book:
    def __init__(self, author, title):
        if not title:
            raise ValueError("Название книги не может быть пустым")
        
        self.author = author
        self.title = title
        self.code = None 

    def tag(self):
        words = self.title.split()
        return [word for word in words if word[0].isupper()]

    def __str__(self):
        name_parts = self.author.split()
        initial = name_parts[0][0] 
        surname = name_parts[1]    
        return f"[{self.code}] {initial}.{surname} '{self.title}'"


class Library:
    book_count = 0

    def __init__(self, number, address):
        self.number = number
        self.address = address
        self.books = [] 

    def __iadd__(self, other):
        if isinstance(other, Book):
            Library.book_count += 1
            other.code = Library.book_count
            
            self.books.append(other)
        return self 

    def __iter__(self):
        return iter(self.books)

lib = Library(1, '51 Some str., NY')
lib += Book('Leo Tolstoi', 'War and Peace')
lib += Book('Charles Dickens', 'David Copperfield')

for book in lib:
    print(book)        
    print(book.tag())  
```
