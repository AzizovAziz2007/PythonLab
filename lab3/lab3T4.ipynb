```python
import re

class StringFormatter:
    def __init__(self, separators=None):
        self.separators = separators if separators else " "

    def _split_to_words(self, s):
        pattern = "[" + re.escape(self.separators) + "]+"
        return re.split(pattern, s)

    def remove_short_words(self, s, n):
        words = self._split_to_words(s)
        filtered_words = [w for w in words if len(w) >= n]
        return " ".join(filtered_words)

    def replace_digits(self, s):
        return re.sub(r'\d', '*', s)

    def insert_spaces(self, s):
        return " ".join(list(s))

    def sort_by_size(self, s):
        words = self._split_to_words(s)
        sorted_words = sorted(words, key=len)
        return " ".join(sorted_words)

    def sort_lexicographical(self, s):
        words = self._split_to_words(s)
        sorted_words = sorted(words)
        return " ".join(sorted_words)

formatter = StringFormatter()
test_str = "your password abcdef12345 is not safe"
print(f"Исходная: {test_str}")
print(f"Без коротких (5): {formatter.remove_short_words(test_str, 5)}")
print(f"Без цифр: {formatter.replace_digits(test_str)}")
```
