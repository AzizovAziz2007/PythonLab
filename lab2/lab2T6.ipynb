```python
import os
import shutil
import argparse
from datetime import datetime, timedelta

def reorganize():
    parser = argparse.ArgumentParser(description="Скрипт для реорганизации файлов")
    parser.add_argument("--source", required=True, help="Путь к исходной директории")
    parser.add_argument("--days", type=int, required=True, help="Количество дней для архивации")
    parser.add_argument("--size", type=int, required=True, help="Максимальный размер для папки Small (в байтах)")
    
    args = parser.parse_args()
    
    source_dir = args.source
    days_limit = args.days
    size_limit = args.size

    if not os.path.exists(source_dir):
        print("Ошибка: Исходная директория не найдена.")
        return

    archive_dir = os.path.join(source_dir, "Archive")
    small_dir = os.path.join(source_dir, "Small")

    now = datetime.now()

    for filename in os.listdir(source_dir):
        file_path = os.path.join(source_dir, filename)

        if not os.path.isfile(file_path):
            continue

        file_mtime = datetime.fromtimestamp(os.path.getmtime(file_path))
        file_size = os.path.getsize(file_path)

        if now - file_mtime > timedelta(days=days_limit):
            if not os.path.exists(archive_dir):
                os.makedirs(archive_dir)
            shutil.move(file_path, os.path.join(archive_dir, filename))
            print(f"Файл {filename} перемещен в Archive")
        
        elif file_size < size_limit:
            if not os.path.exists(small_dir):
                os.makedirs(small_dir)
            shutil.move(file_path, os.path.join(small_dir, filename))
            print(f"Файл {filename} перемещен в Small")

if __name__ == "__main__":
    reorganize()
```
