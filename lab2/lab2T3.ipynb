```python
import os

music_dir = 'lab2T3/music_folder'
list_file = 'lab2T3/songs.txt'

rename_map = {}

try:
    with open(list_file, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            
            dot_pos = line.find('.')      
            bracket_pos = line.find('[')  
            
            if dot_pos != -1 and bracket_pos != -1:
                new_full_name = line[:bracket_pos].strip()
                
                old_base_name = line[dot_pos + 1 : bracket_pos].strip()
                
                rename_map[old_base_name] = new_full_name

    files_in_dir = os.listdir(music_dir)
    
    for filename in files_in_dir:
        base_name, extension = os.path.splitext(filename)
        
        if extension.lower() == '.mp3' and base_name in rename_map:
            new_name = rename_map[base_name] + extension
            
            old_path = os.path.join(music_dir, filename)
            new_path = os.path.join(music_dir, new_name)
            
            os.rename(old_path, new_path)
            print(f"Готово: {filename} -> {new_name}")

    print("\nВсе доступные файлы обновлены!")

except FileNotFoundError:
    print("Ошибка: Не найден файл со списком или указанная папка!")
except Exception as e:
    print(f"Произошла ошибка: {e}")
```
