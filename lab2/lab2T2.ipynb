```python
import os
import hashlib

def get_file_hash(filepath):
    hasher = hashlib.md5()
    try:
        with open(filepath, 'rb') as f:
            while True:
                data = f.read(8192) 
                if not data:
                    break
                hasher.update(data)
        return hasher.hexdigest()
    except:
        return None

root_dir = input("Введите путь к директории для поиска дубликатов: ")

files_by_hash = {}

for root, dirs, files in os.walk(root_dir):
    for filename in files:
        full_path = os.path.join(root, filename)
        
        file_hash = get_file_hash(full_path)
        
        if file_hash:
            if file_hash not in files_by_hash:
                files_by_hash[file_hash] = []
            files_by_hash[file_hash].append(full_path)

print("\nОбнаруженные дубликаты:")
found_duplicates = False

for file_hash, paths in files_by_hash.items():
    if len(paths) > 1:
        found_duplicates = True
        print(f"\nГруппа дубликатов (хеш {file_hash}):")
        for p in paths:
            print(f"  -> {p}")

if not found_duplicates:
    print("Дубликатов не найдено.")
```
